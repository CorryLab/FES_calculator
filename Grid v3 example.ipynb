```python
import Data_Tagger as dt
import Grid_2 as G
from glob import glob
import numpy as np
from fnmatch import fnmatch
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

With the grid manager object, grids are labelled and organised automatically according to a specified scheme
A scheme is an object which contains a list of attribute names that pair up to fields in a filepath
The python object uploaded with that file path is labelled with those attributes. 
See Data_tagger.py for additional details 


```python
# Example run through
# Define a scheme for the labelling of flooding MD directories
# An example directory name is:

# FloodRun_dd2-rep-2_CQH_n-20_rep-5/

floodrun_scheme = dt.Scheme(['_', 'isoform', 'ligand', 'nmols', 'rep'],
                            split_list=['rep', 'nmols'],
                            ignore_attr='_')
# underscores are ignored and not added as attributes to the object
# attr in the split list are split on split_delim kwarg and right most value kept
# The GridManager requires that there be a ligand attr in the list in order to define the ligand selection to use
# The attr list syntax gets fancier, see the OG doc
```

```python
# Specify the directories that you want grids calculated for
dirs = sorted(glob('FloodRun_dd2-rep-?_??H_n-20_rep-*'))

# The generic structure file name within each dir
structure_file = 'mdeq_Prot-CQH-only.gro'

# The generic trajectory file name within each dir
trajectory_file = 'trjcat_centered-Prot_Prot-CQH-only.xtc'

# resolution of the grid
resolution = 1

# Define the GridManager object
Grids = G.GridManager(floodrun_scheme, dirs, structure_file, trajectory_file, resolution)
```

```python
# The GridManager object has a number of convenient functions for selecting subsets of the data
Grids.get_objects(ligand=['QNH'], rep=['2'])
```

```python
# Or values across the uploaded objects can be gotten
Grids.get_values('ligand', unique=True)
```

```python
# Run all grids at once with:
Grids.run(peratom=True)
```

```python
# A function which transforms a tuple of ints that selects regions of a 3d array, 
# and gives the needed coords, width, height to draw the equiovalent selected area with a plt. rectangle
# The drawn rectangle is only valid when drawing over the y
def selection_to_box_dim(x1, x2, y1, y2, z1, z2):
    
    xz = (x1, z1)
    height = z2 - z1
    width = x2 - x1
    
    return xz,  width, height,
```

```python
# Plotting stuff



xyz='xyz'
# select bulk solution slice
selection1 = (0, 91, 0, 91, 85, 90 )
# select binding cavity
selection2 = (30, 70, 30,70, 40,70)


levels = np.arange(-15, 1, 1)

# Complex selections can be performed, or just loop through every grid object in Grid.objs

# Can also perform sorting on Grids.objs with lambda function key operating on object atrributes

# e.g. sort from highest to lowest free energies
# Grids.objs.sort(key=lambda o: np.nanmin(o.fe_grid))


for grid in Grids.get_objects(ligand=['CQH']):

    
    
    fig, axes = plt.subplots(1,3, constrained_layout=True, figsize=(15, 5))
    
    fig.suptitle(grid.tag)
    
    
    for i, ax in enumerate(axes):

        # Calculate and plot FES of drug
        meanfes = np.nanmean(grid.fe_grid , axis=i)
        meanfes = np.rot90(meanfes, k=-1)
        meanfes = np.flip(meanfes, axis=1)
        
        
        cs = ax.contourf(meanfes,   )

       
        ax.set_title(xyz[i])
        colorbar = fig.colorbar(cs, ax=ax)
        
        
#         ax.set_ylim(0,100)
#         ax.set_xlim(0,91)
        
        
        # Drawing of the binding cavity and bulk selections
        
        bc_xz, bc_width, bc_height = selection_to_box_dim(*selection2)
        bulk_xz, bulk_width, bulk_height = selection_to_box_dim(*selection1 )
    
    
        # Draw a rectangle on the graph
        BC_selection = patches.Rectangle(bc_xz, bc_width, bc_height, linewidth=1, edgecolor='b', facecolor='none')
        bulk_selection = patches.Rectangle(bulk_xz, bulk_width, bulk_height, linewidth=1, edgecolor='r', facecolor='none', )

        # Add the patch to the Axes
        ax.add_patch(BC_selection)
        ax.add_patch(bulk_selection)
        
        
            
    plt.show()
    
```
